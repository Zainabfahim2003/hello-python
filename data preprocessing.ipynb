```python
import pandas as pd

# 🧠 Sample dataframes loaded here (assume you've already loaded them)
datasets = {
    'df1_indicators': df1,
    'df2_soil_data': df2,
    'df3_environmental_data': df3,
    'df4_soil_data_again': df4
}

# 📌 Function to print potential merge columns between each pair
def find_mergeable_columns(datasets):
    keys = list(datasets.keys())
    for i in range(len(keys)):
        for j in range(i+1, len(keys)):
            dfA = datasets[keys[i]]
            dfB = datasets[keys[j]]
            
            common_cols = set(dfA.columns).intersection(set(dfB.columns))
            same_dtype_cols = [col for col in common_cols if dfA[col].dtype == dfB[col].dtype]
            
            print(f"🔗 Mergeable columns between **{keys[i]}** and **{keys[j]}**:")
            if same_dtype_cols:
                for col in same_dtype_cols:
                    print(f"  ✅ {col} (dtype: {dfA[col].dtype})")
            else:
                print("  ❌ No directly mergeable columns (same name + dtype)")
            print("-" * 60)

# 🔍 Run the function
find_mergeable_columns(datasets)
```
